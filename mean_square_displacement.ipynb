### Generate data

```python
# The Wiener process parameter.
delta = 0.25
# Total time.
time = 10.0
# Number of steps.
nstep = 20000
# Time step size
dt = time/nstep
# Initial values of x.
xy = np.empty((2,nstep+1))
xy[:, 0] = 0.0

data = brownian(xy[:,0], nstep, dt, delta, out=xy[:,1:])
data.shape
```

### Plot the data

```python
plot_brownian(data)
```

### Compute MSD

```python
msd_result = compute_msd(data.T)
msd_result.shape
```

Plot MSD

```python
len(np.arange(len(msd_result)))
```
